```python
import matplotlib.pyplot as plt
import numpy as np

#Data for bar chart
categories = ['Electronics', 'Clothing', 'Home']
sales = [300, 450, 200]

#Data for line plot (example monthly trend)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
monthly_sales = [200, 250, 300, 280, 350]

#Create Figure
plt.figure(figsize=(10, 4))

#Subplot 1 (Bar Chart)
plt.subplot(1, 2, 1)
plt.bar(categories, sales)
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

#Subplot 2 (Line Plot)
plt.subplot(1, 2, 2)
plt.plot(months, monthly_sales, marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

#Prevent overlapping
plt.tight_layout()

#Show plots
plt.show()
```
