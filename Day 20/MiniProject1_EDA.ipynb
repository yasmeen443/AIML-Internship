```python
df.hist(figsize=(10,8))
plt.show()
```

```python
sns.countplot(x=df.select_dtypes(include='object').columns[0], data=df)
plt.show()
```

```python
sns.boxplot(y=df.select_dtypes(include='number').columns[0], data=df)
plt.show()
```

```python
sns.scatterplot(
    x=df.select_dtypes(include='number').columns[0],
    y=df.select_dtypes(include='number').columns[1],
    data=df
)
plt.show()
```

```python
sns.boxplot(
    x=df.select_dtypes(include='object').columns[0],
    y=df.select_dtypes(include='number').columns[0],
    data=df
)
plt.show()
```
