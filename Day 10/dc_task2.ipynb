```python
import pandas as pd

data = {
    "Price": ["$100", "$250", "$300", "$150"],
    "Date": ["2025-01-01", "2025-01-05", "2025-01-10", "2025-01-15"]
}

df = pd.DataFrame(data)

print(df)
print(df.dtypes)
```

```python
df["Price"]=df["Price"].str.replace("$","",regex=False).astype(float)
```

```python
df["Price"]
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```
